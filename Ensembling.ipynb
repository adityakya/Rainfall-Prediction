# ENSEMBLING

So our last model did well. Now we will try some ensembling methods.Ensembling is a general term for combining many classifiers by averaging or voting.Generally, ensembles of classifiers perform better than single classifiers, and the averaging process allows for moregranularity of choice in the bias-variance tradeoff.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base_model1=pd.read_csv("hybrid.csv")
base_model2=pd.read_csv("Reflectivity.csv")
```

```python
cols=list(base_model1.columns)
cols.remove("Id")
print(len(cols))
```

```python
ensemble1=base_model1[cols]/2+base_model2[cols]/2
ensemble1.head(10)
```

```python
ensemble1=pd.concat([base_model1['Id'],ensemble1],axis=1)
ensemble1.head(10)
```

```python
ensemble1.to_csv("Average_ensemble.csv",index=False)
print("Done")
```

```python
ensemble2=base_model1[cols]*0.3+base_model2[cols]*0.7
ensemble2.head(10)
```

```python
ensemble2=pd.concat([base_model1['Id'],ensemble2],axis=1)
ensemble2.head(10)
```

```python
ensemble2.to_csv("Voting_0.7_Qc.csv",index=False)
```
